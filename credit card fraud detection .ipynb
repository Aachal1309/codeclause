# Creating dependent and independent variables


```python
X= new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler= StandardScaler()
```

```python
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train= Scaler.fit_transform(X_train)
```

```python
X_test= Scaler.fit_transform(X_test)
```

```python
print( X.shape ,X_train.shape , X_test.shape)
```

```python
print( Y.shape ,Y_train.shape , Y_test.shape)
```

# Logistic Regression


```python
clf= LogisticRegression()
```

```python
#Training Model
```

```python
clf.fit(X_train,Y_train)
```

```python
X_train_predict =clf.predict(X_train)
training_data_accuracy=accuracy_score(X_train_predict ,Y_train)
```

```python
print("the accuracy score on training dataset is ", training_data_accuracy)
```

```python
X_test_predict =clf.predict(X_test)
test_data_accuracy=accuracy_score(X_test_predict ,Y_test)
```

```python
print("the accuracy score on test dataset is ", test_data_accuracy)
```
